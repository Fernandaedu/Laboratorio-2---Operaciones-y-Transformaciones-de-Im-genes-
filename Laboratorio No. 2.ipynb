## Problema # 2
Implemente una funcin para cada una de las transformaciones vistas en clase: negativa, lineal, logartmica y exponencial.

```python
# Cargar imagen

img = cv2.imread("imagen.jpeg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img, cmap='gray')
```

```python
# Funcin 'Negativa', 'Lineal', 'Logaritmica', 'Exponecial'

def image_trans(image, transformation):

  # Dimensiones de la imagen
  alto = image.shape[0]
  ancho = image.shape[1]

  image_trans = np.zeros((alto, ancho), np.uint8)

  images_list = []
  titles = ['ORIGINAL']

  images_list.append(image)
  titles.append(transformation)

  for i in range(alto):
      for j in range(ancho):
          if transformation == 'NEGATIVA':
            image_trans[i][j] = 255 - image[i][j]

          if transformation == 'LINEAL':
            image_trans[i][j] = 1.5 * image[i][j] + 10

          if transformation == 'LOGARITMICA':
            image_trans[i][j] = 40 * np.log(1 + image[i][j])

          if transformation == 'EXPONENCIAL': 
            image_trans[i][j] = 2 * np.power(image[i][j], 0.9)

  images_list.append(image_trans)

  plt.figure(figsize=(15,15))
  for i in range(len(images_list)):
    plt.subplot(1, len(images_list), i + 1)
    plt.imshow(images_list[i], cmap='gray')
    plt.title("IMAGE {}".format(titles[i]), fontdict={"fontsize":16})
  plt.show()
     
```

```python
trans_dict = {1:'NEGATIVA',
                       2:'LINEAL',
                       3:'LOGARITMICA',
                       4:'EXPONENCIAL'}
```

```python
# Transformacin Negativa

image_trans(img, trans_dict[1])
```

```python
# Transformacin Lineal

image_trans(img, trans_dict[2])
```

```python
# Transformacin Logaritmica

image_trans(img, trans_dict[3])
```

```python
# Transformacin Exponencial

image_trans(img, trans_dict[4])
```

## Problema # 3
Desarrolle una funcin que dada dos imgenes a color, realice las operaciones SUMA, RESTA, AND,OR , XOR con ambas imgenes recuerde que para realizar las operaciones lgicas es necesario pasar la imagen a escala de grises y luego binarizarla

```python
# Cargar imagen 1

imagen1=cv2.imread("imagen1.jpeg")
imagen1=cv2.cvtColor(imagen1,cv2.COLOR_BGR2RGB)
plt.imshow(imagen1)
plt.show()
```

```python
imagen2=cv2.imread("imagen2.jpeg")
imagen2=cv2.cvtColor(imagen2,cv2.COLOR_BGR2RGB)
plt.imshow(imagen2)
plt.show()
```

```python

# Suma

img_ADD = cv2.add(imagen1, imagen2)
img_ADD = cv2.cvtColor(img_ADD, cv2.COLOR_BGR2RGB)
plt.imshow(img_ADD)
```

```python
# Resta

img_DIFF = cv2.absdiff(imagen1, imagen2)
img_DIFF = cv2.cvtColor(img_DIFF, cv2.COLOR_BGR2RGB)
plt.imshow(img_DIFF)
```

```python
# AND

img_AND = cv2.bitwise_and(imagen1, imagen2)
img_AND = cv2.cvtColor(img_AND, cv2.COLOR_BGR2RGB)
plt.imshow(img_AND)
```

```python
# OR

img_OR = cv2.bitwise_or(imagen1, imagen2)
img_OR = cv2.cvtColor(img_OR, cv2.COLOR_BGR2RGB)
plt.imshow(img_OR)
```

```python
# XOR

img_XOR = cv2.bitwise_xor(imagen1, imagen2)
img_XOR = cv2.cvtColor(img_XOR, cv2.COLOR_BGR2RGB)
plt.imshow(img_XOR)
```

```python
# CANNY

image = cv2.Canny(imagenvol, threshold1=100, threshold2=200)

cv2.imshow('Original', imagenvol)
cv2.imshow('Edges (Canny)', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
plt.imshow(image)
```

```python
# SOBEL

ddepth = cv2.CV_16S
ksize = 3
dx = 2
dy = 2

image = cv2.Sobel(imagenvol, ddepth=ddepth, dx=dx, dy=dy, ksize=ksize)
plt.imshow(image)
```

```python
# LAPLACIANO

ddepth = cv2.CV_16S
ksize = 3

image = cv2.Laplacian(imagenvol, ddepth=ddepth, ksize=ksize)
plt.imshow(image)
```
